```python
import pandas as pd

df = pd.read_csv("notebooks/data/taxi_gold_enriched.csv")
# aperçu
print(df.head())

# dimensions
print("Shape:", df.shape)

# informations sur les variables
print(df.info())

# statistiques descriptives
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.histplot(df["Duree_min"], bins=30)
plt.title("Distribution de la durée de taxi")
plt.xlabel("Durée (minutes)")
plt.ylabel("Fréquence")
plt.show()
```

```python
plt.figure()
sns.histplot(df["Consommation_totale_kg"], bins=30)
plt.title("Distribution de la consommation carburant")
plt.xlabel("Consommation (kg)")
plt.ylabel("Fréquence")
plt.show()
```

```python
plt.figure()
sns.scatterplot(
    x="Duree_min",
    y="Consommation_totale_kg",
    data=df
)

plt.title("Consommation en fonction de la durée de taxi")
plt.xlabel("Durée (min)")
plt.ylabel("Consommation (kg)")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Matrice de corrélation")
plt.show()
```

```python
corr_val = df["Duree_min"].corr(df["Consommation_totale_kg"])
print("Corrélation durée / consommation :", corr_val)
```
