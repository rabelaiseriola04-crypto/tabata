## Extraction

```python
import pandas as pd
import h5py
import os

# Liste des colonnes cibles (identifiées précédemment)
target_cols = ['ALT [ft]', 'M [Mach]', 'Q_1 [lb/h]', 'Q_2 [lb/h]', 
               'N1_1 [% rpm]', 'N1_2 [% rpm]', 'N2_1 [% rpm]', 'N2_2 [% rpm]', 
               'TLA_1 [deg]', 'TLA_2 [deg]']

def clean_extraction_v4(folder_path):
    final_list = []
    files = [f for f in os.listdir(folder_path) if f.endswith('.h5')]
    
    for file_name in files:
        with h5py.File(os.path.join(folder_path, file_name), "r") as f:
            for record_name in f.keys():
                df = pd.read_hdf(os.path.join(folder_path, file_name), key=record_name)
                
                # 1. Sélection et nettoyage immédiat
                # On ne garde que si les colonnes existent
                available_cols = [c for c in target_cols if c in df.columns]
                if len(available_cols) < len(target_cols):
                    continue 
                
                df_slim = df[available_cols].copy()
                
                # 2. Filtre Taxi (ALT < 50 et on évite les valeurs folles < -5)
                mask = (df_slim['ALT [ft]'] < 50) & (df_slim['ALT [ft]'] > -5)
                df_taxi = df_slim[mask].copy()
                
                if not df_taxi.empty:
                    df_taxi['flight_id'] = file_name
                    df_taxi['record_id'] = record_name
                    final_list.append(df_taxi)

    # L'astuce est ici : concaténation verticale pure
    return pd.concat(final_list, axis=0, ignore_index=True)

# Relance l'extraction
df_clean = clean_extraction_v4("tabata/notebooks/data/dfdr1000")
df_clean.to_csv("tabata/notebooks/data/taxi_clean_final.csv", index=False)
```

## Croisement

```python
import pandas as pd

# 1. Chargement de la base extraite au point précédent
df = pd.read_csv("tabata/notebooks/data/taxi_clean_final.csv")

def perform_data_crossing(df_input):
    df_crossed = df_input.copy()
    
    # --- A. Croisement des flux de consommation (Moteur 1 + Moteur 2) ---
    # On crée une vision "Avion" globale
    df_crossed['Q_Total [lb/h]'] = df_crossed['Q_1 [lb/h]'] + df_crossed['Q_2 [lb/h]']
    
    # --- B. Croisement des commandes pilotes (TLA Moyen) ---
    # On croise les deux manettes pour avoir l'intention moyenne de poussée
    df_crossed['TLA_Avg [deg]'] = (df_crossed['TLA_1 [deg]'] + df_crossed['TLA_2 [deg]']) / 2
    
    # --- C. Croisement des régimes moteurs (N1 Moyen) ---
    # Utile pour vérifier la symétrie de la poussée au taxi
    df_crossed['N1_Avg [%]'] = (df_crossed['N1_1 [% rpm]'] + df_crossed['N1_2 [% rpm]']) / 2

    # --- D. Calcul du Delta de poussée (Optionnel pour le nettoyage) ---
    # Permet de détecter si un moteur est beaucoup plus sollicité que l'autre
    df_crossed['N1_Delta [%]'] = abs(df_crossed['N1_1 [% rpm]'] - df_crossed['N1_2 [% rpm]'])
    
    return df_crossed

# 2. Exécution du croisement
df_final_crossed = perform_data_crossing(df)

# 3. Sauvegarde pour l'étape suivante (Calcul d'indicateurs)
df_final_crossed.to_csv("tabata/notebooks/data/taxi_crossed_data.csv", index=False)

print("✅ Croisement terminé !")
print(f"Colonnes après croisement : {df_final_crossed.columns.tolist()}")
print(df_final_crossed[['Q_Total [lb/h]', 'TLA_Avg [deg]', 'N1_Avg [%]']].head())
```

## Calcul des indicateurs

```python
import pandas as pd

# 1. Charger la base de données croisée (celle avec Q_Total)
df_crossed = pd.read_csv("tabata/notebooks/data/taxi_crossed_data.csv")

def generate_taxi_indicators(df):
    # --- A. Transformation Physique (lb/h en kg par seconde) ---
    # Chaque ligne valant 1s, on divise le débit horaire par 3600 puis on convertit en kg
    df['Conso_kg_s'] = (df['Q_Total [lb/h]'] / 3600) * 0.453592
    
    # --- B. Agrégation par Segment (Le Résumé) ---
    # On groupe par vol et par record pour calculer les totaux et moyennes
    summary = df.groupby(['flight_id', 'record_id']).agg(
        Duree_sec=('ALT [ft]', 'count'),             # Nombre de lignes = Durée en secondes
        Consommation_totale_kg=('Conso_kg_s', 'sum'), # Somme des kg consommés sur tout le taxi
        TLA_moyen=('TLA_Avg [deg]', 'mean'),         # Intensité moyenne de la manette
        N1_moyen=('N1_Avg [%]', 'mean'),             # Régime moteur moyen
        Vitesse_max_Mach=('M [Mach]', 'max')          # Vitesse de pointe au sol
    ).reset_index()
    
    # --- C. Création d'indicateurs "Métier" ---
    # Conversion de la durée en minutes pour plus de lisibilité
    summary['Duree_min'] = summary['Duree_sec'] / 60
    
    # Efficacité : kg consommés par minute de taxi
    summary['Efficacite_kg_min'] = summary['Consommation_totale_kg'] / summary['Duree_min']
    
    return summary

# 2. Exécution du calcul
taxi_final_summary = generate_taxi_indicators(df_crossed)

# 3. Nettoyage final : On retire les records "fantômes" (moins de 1 min ou conso nulle)
# Cela évite de polluer l'IA avec des records où l'avion n'a pas vraiment bougé
taxi_final_summary = taxi_final_summary[(taxi_final_summary['Duree_min'] > 1) & 
                                        (taxi_final_summary['Consommation_totale_kg'] > 0.1)]

# 4. Sauvegarde de la Base de Données Finale (Le Graal du projet)
taxi_final_summary.to_csv("tabata/notebooks/data/taxi_final_summary.csv", index=False)

print("✅ Jalon 3 (Calcul d'indicateurs) terminé !")
print(f"Nombre de trajets de taxi identifiés : {len(taxi_final_summary)}")
print(taxi_final_summary.head())
```

## Nettoyage

```python
# --- ÉTAPE 3D : NETTOYAGE FINAL ---

import pandas as pd

# 1. Chargement du résumé calculé à l'étape précédente
df_summary = pd.read_csv("tabata/notebooks/data/taxi_final_summary.csv")

# 2. Définition des filtres de cohérence (Physique Aéronautique)
# - Au moins 2 minutes (pour exclure les simples tests moteur)
# - Au moins 10 kg (pour exclure les erreurs de capteurs Q=0)
# - Vitesse < 0.25 Mach (pour exclure les décollages accidentels)
mask = (df_summary['Duree_min'] >= 2.0) & \
       (df_summary['Consommation_totale_kg'] >= 10.0) & \
       (df_summary['Vitesse_max_Mach'] <= 0.25)

df_gold = df_summary[mask].copy()

# 3. Calcul de la perte de données (pour ton Powerpoint)
initial = len(df_summary)
final = len(df_gold)
perte = initial - final

print(f"--- RÉSULTAT DU NETTOYAGE (J3-D) ---")
print(f"Trajets analysés au départ : {initial}")
print(f"Trajets supprimés (car aberrants) : {perte}")
print(f"Trajets restants pour l'IA (Dataset Gold) : {final}")

print("\n--- TABLEAU DE BORD DE QUALITÉ (À METTRE DANS LE POWER) ---")
# On affiche les stats clés pour prouver la cohérence
print(df_gold[['Duree_min', 'Consommation_totale_kg', 'Efficacite_kg_min']].describe())

# 4. Sauvegarde du dataset final "Gold" pour le Jalon 4 et 5
df_gold.to_csv("tabata/notebooks/data/taxi_gold_dataset.csv", index=False)
```
